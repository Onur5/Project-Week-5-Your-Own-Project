# Are we still exploiting Africa?
### Agriculture: A simple analysis

In recent years, there were many articels about the EU exploiting Africa today. Especially in terms of international trade contracts for agricultural products the business circumstances seem to unbalanced. African countries are general disadvantaged in international food business.
This project aims to analysis the trade flow for agricultural products between the EU and the sub-sahara African region.
