## 1. Balance Sheets

As an inital step the agricultural balance sheets of the regions will be analysed. The dataset is obtained from the Food and Agriculture Organization (FAO) of the UN, who provide a database with the balance sheet of agricultural products for most countries ([FAO-STAT](http://www.fao.org/faostat/en/#data/FBS)).
The main interest lies here in the net exports(export-import).
For the analysis the total period available will be used (1960 - 2013).

At this point it is important to note that on the database only item groups were selected that are comparable between the regions. For example the production and trade of silk or cotton was not covered as it is not a product of the european agriculture. Of course this judgement is subject to some extent. Nevertheless some compromises had to be made at this point to fit the scope of the project.
Product groups covered are "Fruits - Excluding Wine", "Eggs", "Sugar & Sweeteners", "Meat", "Vegetables", "Cereals - Excluding Beer", "Alcoholic Beverages"	and "Milk - Excluding Butter".

For organisational issues the balance sheet of the EU will be analysed first:

```python
food_bs = pd.read_csv("../data/FAOSTAT_Europe.csv")
```

```python
food_bs.head()
```

First an initial data cleaning to obtain the information needed

```python
import_bs = food_bs[food_bs["Element"]=="Import Quantity"].rename(columns={'Value':'Import'})
del import_bs["Domain Code"]
del import_bs["Domain"]
```

or rather "balance of trade" ("BOT" in the dataframe)

```python
df_food_bs["BOT"] = df_food_bs["Export"] - df_food_bs["Import"]
```

```python
df_food_bs.head()
```

## 2. Direct Trade: EU vs Africa

As the previous graphs only depict the agricultural balance sheet of the two regions, it is not possible to make a statement about trade relation between the countries. This is the reason why the international agricultural product trade between the two countries will be analyzed in the following.
The dataset was obtained again from the FAO of the UN, who also provide a "detailed trade matrix" of countries for agricultural products ([link](http://www.fao.org/faostat/en/#data/TM)).
Again the total available period will be used for the analysis (1986 - 2016).
The FAO provides information on the import and export quantity for each country/region. However, other than the food balance sheet database, also the import and export quantity is available in this database, which provides additional information.

Eventhough an inbalanced is also expected for all the prior compared agricultural products (mainly milk-products, as these are mentioned primaraly in the media), due to the time restirctions of the project, the direkt trade between the EU and sub-sahara Africa will be analyzed for Cereals and Meat only

First the the direct trade for cereals will be cleaned and analyzed and afterwards the data for the direct trade of meat. In a last step the dataframes will be merged and represented in a combined graph. 

```python
trade_matrix = pd.read_csv("../data/EUvsAfrica.csv")
trade_matrix
```

As the underlying dataset doesnt reperesent the aggregated regions (EU and sub-sahara Africa), but all the countries in this regions, first the dataframe have to be adjusted. As the reporter countries are only members of the EU and the partner countries only countries of sub-sahara Africa, a simple grouping and summing of the data will be applied here. <br> <br>
However, as the dataset provides information on the import and export (quantity and value), each observation (each row) represents either the import quantity, import value, export quantity or export value (for a specific reporter "EU-country" with a specific partner "African-country" in a specific year). <br> The column "Element" describes which information the observation contains. <br><br>
Thus, in the following the data will be cleaned first from null values (as some countries have no trade in some years) and afterwards the data is grouped by 'Year' and 'Element' and summed. This allows us to obtain the true values for the regions.

```python
trade_matrix= trade_matrix[trade_matrix.Value != 0][["Reporter Countries", "Partner Countries", "Element", "Year", "Value"]]
trade_matrix= trade_matrix.groupby(by=["Element", "Year"]).sum().reset_index()
```

```python
set(trade_matrix["Element"])
```

To have a better overview, all of these values will be gathered into one single observation (so that each row holds the value of import and export (quantity and value): <br>
For this the dataset will be splitted into four different datasets (import quantity, import value, export quantity or export value respectively) first.

```python
df_trade_import_q = trade_matrix[trade_matrix["Element"]=="Import Quantity"].rename(columns={'Value':'Import_Q'})
del df_trade_import_q["Element"]
df_trade_export_q = trade_matrix[trade_matrix["Element"]=="Export Quantity"].rename(columns={'Value':'Export_Q'})
del df_trade_export_q["Element"]
df_trade_import_v = trade_matrix[trade_matrix["Element"]=="Import Value"].rename(columns={'Value':'Import_V'})
del df_trade_import_v["Element"]
df_trade_export_v = trade_matrix[trade_matrix["Element"]=="Export Value"].rename(columns={'Value':'Export_V'})
del df_trade_export_v["Element"]
```

Afterwards the datasets will be merged.

```python
df_trade_step1 = df_trade_import_q.merge(df_trade_export_q, how='left', left_on=["Year"], right_on=["Year"])
df_trade_step2 = df_trade_step1.merge(df_trade_import_v, how='left', left_on=["Year"], right_on=["Year"])
df_trade_matrix = df_trade_step2.merge(df_trade_export_v, how='left', left_on=["Year"], right_on=["Year"])
```

Calculation of "net exports" - VARIABLE OF INTEREST <br> <br>As before the variable of interest is the net export ('BOT' in the dataframe) and will be calculated for the exported/imported quantities ('BOT_Q' in the dataframe) and for the exported/imported values ('BOT_V' in the dataframe).

```python
df_trade_matrix["BOT_Q"] = df_trade_matrix["Export_Q"] - df_trade_matrix["Import_Q"]
df_trade_matrix["BOT_V"] = df_trade_matrix["Export_V"] - df_trade_matrix["Import_V"]
```

```python
df_trade_matrix.head()
```

As mentioned before, the same adjustments are also done for the meat trade between the countries.

<br> Merging of countries to regions

```python
trade_matrix_meat = trade_matrix_meat[trade_matrix_meat.Value != 0][["Reporter Countries", "Partner Countries", "Element", "Year", "Value"]]
trade_matrix_meat = trade_matrix_meat.groupby(by=["Element", "Year"]).sum().reset_index()
```

<br>
Change of the dataset structure to represent import and export (quantity and value) in one observation as before

```python
df_trade_import_meat_q = trade_matrix_meat[trade_matrix_meat["Element"]=="Import Quantity"].rename(columns={'Value':'Import_Q'})
del df_trade_import_meat_q["Element"]
df_trade_export_meat_q = trade_matrix_meat[trade_matrix_meat["Element"]=="Export Quantity"].rename(columns={'Value':'Export_Q'})
del df_trade_export_meat_q["Element"]
df_trade_import_meat_v = trade_matrix_meat[trade_matrix_meat["Element"]=="Import Value"].rename(columns={'Value':'Import_V'})
del df_trade_import_meat_v["Element"]
df_trade_export_meat_v = trade_matrix_meat[trade_matrix_meat["Element"]=="Export Value"].rename(columns={'Value':'Export_V'})
del df_trade_export_meat_v["Element"]
```

```python
df_trade_meat_step1 = df_trade_import_meat_q.merge(df_trade_export_meat_q, how='left', left_on=["Year"], right_on=["Year"])
df_trade_meat_step2 = df_trade_meat_step1.merge(df_trade_import_meat_v, how='left', left_on=["Year"], right_on=["Year"])
df_trade_meat_matrix = df_trade_meat_step2.merge(df_trade_export_meat_v, how='left', left_on=["Year"], right_on=["Year"])
```

<br>
Calculation of "net exports" - VARIABLE OF INTEREST ('BOT')

```python
df_trade_meat_matrix["BOT_Q"] = df_trade_meat_matrix["Export_Q"] - df_trade_meat_matrix["Import_Q"]
df_trade_meat_matrix["BOT_V"] = df_trade_meat_matrix["Export_V"] - df_trade_meat_matrix["Import_V"]
```
